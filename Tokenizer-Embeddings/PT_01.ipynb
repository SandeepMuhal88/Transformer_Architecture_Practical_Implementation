```python
import torch
import torch.nn as nn

vocab_size = 30522   # BERT vocab
embed_dim = 768

embedding = nn.Embedding(vocab_size, embed_dim)

token_ids = torch.tensor([[101, 2054, 2003, 2115, 2171, 102]])
embeddings = embedding(token_ids)

print(embeddings.shape)  # [batch, seq_len, embed_dim]
```
